```python
from scipy.stats import weibull_min as weibull
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rv = weibull(c=0.9, loc=3.0)
```

```python
x = np.arange(0.01, 10.0, 0.01)
```

```python
f_x = rv.pdf(x)
```

```python
plt.plot(x,f_x)
```

```python
s_x = rv.sf(x)
```

```python
plt.plot(x, s_x)
```

```python
plt.plot(x, f_x/s_x)
```

```python
rv = weibull(c=1.3, scale=1000000000.0)
```

```python
data = rv.rvs(size=500)
```

```python
out_string = ""
for d in data:
    x=int(np.ceil(d))
    out_string += str(x) + "\n"

with open("test_data.txt", 'w') as f:
    f.write(out_string)
```
